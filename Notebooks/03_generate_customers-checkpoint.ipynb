```python
from sqlalchemy import create_engine
import pandas as pd

DB_USER = "postgres"
DB_PASSWORD = "Fardeen123"      # Replace with your password
DB_HOST = "localhost"
DB_PORT = "5432"
DB_NAME = "b2b_churn_analytics"

engine = create_engine(
    f"postgresql+psycopg2://{DB_USER}:{DB_PASSWORD}@{DB_HOST}:{DB_PORT}/{DB_NAME}"
)
```

```python
import random
from faker import Faker
import pandas as pd

fake = Faker()

industries = [
    "Healthcare",
    "Banking",
    "Retail",
    "Manufacturing",
    "Education",
    "Logistics",
    "Technology",
    "Hospitality"
]

countries = [
    "United Arab Emirates",
    "Saudi Arabia",
    "Qatar",
    "Oman",
    "Bahrain",
    "India",
    "Singapore",
    "United Kingdom"
]

company_sizes = [
    "Small",
    "Medium",
    "Enterprise"
]

customers = []

for _ in range(100):
    customers.append({
        "company_name": fake.company(),
        "industry": random.choice(industries),
        "country": random.choice(countries),
        "company_size": random.choice(company_sizes),
        "annual_revenue": round(random.uniform(50000, 5000000), 2),
        "signup_date": fake.date_between(start_date="-5y", end_date="today"),
        "account_status": random.choice(["Active", "Active", "Active", "Cancelled"]),
        "sales_rep_id": random.randint(1, 20)
    })

customers_df = pd.DataFrame(customers)

customers_df.head()
```

```python
#Validate the Data

customers_df.info()

customers_df.isnull().sum()
```

```python
pd.read_sql("SELECT version();", engine)
```

```python
customers_df.to_sql(
    "customers",
    engine,
    if_exists="append",
    index=False
)
```

```python
pd.read_sql("""
SELECT
    customer_id,
    company_name,
    industry,
    country,
    annual_revenue,
    sales_rep_id
FROM customers
LIMIT 10;
""", engine)
```
