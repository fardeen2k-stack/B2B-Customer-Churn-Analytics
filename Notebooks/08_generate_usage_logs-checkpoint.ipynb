```python
import random
import pandas as pd
from datetime import timedelta
from faker import Faker

fake = Faker()
```

```python
customers = pd.read_sql("""
SELECT customer_id
FROM customers;
""", engine)

customers.head()
```

```python
from datetime import timedelta

usage_logs = []

for customer in customers["customer_id"]:

    start_date = pd.Timestamp("2025-06-01")

    for day in range(30):

        usage_logs.append({

            "customer_id": customer,

            "usage_date": start_date + timedelta(days=day),

            "active_users": random.randint(1,50),

            "login_count": random.randint(1,25),

            "api_calls": random.randint(50,5000),

            "storage_used_gb": round(random.uniform(5,500),2),

            "session_duration_minutes": random.randint(5,240)

        })

usage_df = pd.DataFrame(usage_logs)

usage_df.head()
```

```python
usage_df.shape
```

```python
usage_df.to_sql(
    "usage_logs",
    engine,
    if_exists="append",
    index=False
)
```
