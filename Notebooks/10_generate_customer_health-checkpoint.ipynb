```python
from sqlalchemy import create_engine
import pandas as pd

DB_USER = "postgres"
DB_PASSWORD = "Fardeen123"      # Replace with your password
DB_HOST = "localhost"
DB_PORT = "5432"
DB_NAME = "b2b_churn_analytics"

engine = create_engine(
    f"postgresql+psycopg2://{DB_USER}:{DB_PASSWORD}@{DB_HOST}:{DB_PORT}/{DB_NAME}"
)
```

```python
import random
import pandas as pd
from faker import Faker

fake = Faker()

customers = pd.read_sql("""
SELECT customer_id
FROM customers;
""", engine)
```

```python
from datetime import timedelta

customer_health = []

for customer in customers["customer_id"]:

    health_score = random.randint(40, 100)

    if health_score >= 80:
        churn_risk = "Low"
    elif health_score >= 60:
        churn_risk = "Medium"
    else:
        churn_risk = "High"

    customer_health.append({

        "customer_id": customer,

        "health_score": health_score,

        "churn_risk": churn_risk,

        "last_login": fake.date_between(
            start_date="-30d",
            end_date="today"
        ),

        "nps_score": random.randint(0, 10)

    })

health_df = pd.DataFrame(customer_health)

health_df.head()
```

```python
health_df.to_sql(
    "customer_health",
    engine,
    if_exists="append",
    index=False
)
```
